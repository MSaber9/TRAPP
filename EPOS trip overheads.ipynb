```python
# print histogram

import numpy as np

n_bins = 6
fig, axes = plt.subplots(nrows=1, ncols=1)
ax0 = axes

c1 = np.ones(500) * -1
c2 = np.ones(500) * -1
c3 = np.ones(500) * -1
c4 = np.ones(500) * -1
c5 = np.ones(500) * -1


c1[0 : len(car200)] = car200
c2[0 : len(car400)] = car400
c3[0 : len(car600)] = car600
c4[0 : len(car800)] = car800
c5[0 : len(car1000)] = car1000

cc = np.stack((c1,c2,c3,c4,c5))
cc = cc.transpose()


#colors = ['#44000d', '#83142c', '#ad1d45']
colors = ['#00a898', '#5ebf8c', '#b6d887' ]
colors = ['#8390A8', '#5A6B8C', '#31466F', '#1B3059', '#162647']
labels = ['200 cars', '400 cars', '600 cars', '800 cars', '1000 cars']
ax0.hist(cc, n_bins, normed=None, histtype='bar', color=colors, label=labels)
#ax0.hist(cc2, n_bins, normed=1, histtype='bar', color='#b6d887', label='car200')
ax0.legend(prop={'size': 10})
ax0.set_title('Trip Overhead of cars 200 till 1000 in a map of Eichstatt')
ax0.set_xlabel('Trip Overheads')
ax0.set_ylabel('Cars')

fig.tight_layout()
plt.show()

print("NOTE ABOUT HISTOGRAM")
print("Ignore the first stack on histogram showing less than 0 overheads. This is a padding on the data required for histogram")
print("---------------------------")
print("x-axis shows the overheads")
print("y-axis shows the no. of cars")
```
