```python
%matplotlib inline  
import matplotlib as mpl
import matplotlib.pyplot as plt
from statistics import mean, median, variance 
plt.rcParams['figure.figsize'] = [10, 5]
import pprint

def get_data_from_file(file_name):
    utilizations = []
    line_no = 0
        
    with open(file_name, 'r') as results:
        for line in results:
            line = line.split(",")
            line[len(line)-1] = line[len(line)-1].replace('\r', '').replace('\n', '')
            line_no = line_no + 1
            if line_no == 1:
                edges = line
            else: 
                utilizations.append([float(u) for u in line[0:]])   
    
    edges_data = {}
    for i in range(len(edges)):
        utilizations_of_edge = []
        for utilization in utilizations:
            utilizations_of_edge.append(utilization[i])
        edges_data[edges[i]] = utilizations_of_edge
    
    return edges_data

# utilization_altruistic = get_data_from_file("results/edges_a0_b0.csv")
# utilization_selfish = get_data_from_file("results/edges_a0_b1.csv")
# utilization_a0_b09 = get_data_from_file("results/edges_a0_b0.9.csv")
# utilization_a0_b05 = get_data_from_file("results/edges_a0_b0.5.csv")

# a1b0 = get_data_from_file("../graph backups/exp-new-utilization/old-utilization-process/a1-b0/streets.csv")
# a1b25 = get_data_from_file("../graph backups/exp-new-utilization/old-utilization-process/a1-b25/streets.csv")
# a1b50 = get_data_from_file("../graph backups/exp-new-utilization/old-utilization-process/a1-b50/streets.csv")
# a1b75 = get_data_from_file("../graph backups/exp-new-utilization/old-utilization-process/a1-b75/streets.csv")
# a1b100 = get_data_from_file("../graph backups/exp-new-utilization/old-utilization-process/a1-b100/streets.csv")


a1b0 = get_data_from_file("../graph backups/exp-new-utilization/new-utilization-process/a1-b0/volume.csv")
a1b25 = get_data_from_file("../graph backups/exp-new-utilization/new-utilization-process/a1-b25/volume.csv")
a1b50 = get_data_from_file("../graph backups/exp-new-utilization/new-utilization-process/a1-b50/volume.csv")
a1b75 = get_data_from_file("../graph backups/exp-new-utilization/new-utilization-process/a1-b75/volume.csv")
a1b100 = get_data_from_file("../graph backups/exp-new-utilization/new-utilization-process/a1-b100/volume.csv")

a1b0_means = {}
a1b25_means = {}
a1b50_means = {}
a1b75_means = {}
a1b100_means = {}

for key, value in a1b0.iteritems():
    a1b0_means[key] = mean(value)
    
for key, value in a1b25.iteritems():
    a1b25_means[key] = mean(value)

for key, value in a1b50.iteritems():
    a1b50_means[key] = mean(value)

for key, value in a1b75.iteritems():
    a1b75_means[key] = mean(value)

for key, value in a1b100.iteritems():
    a1b100_means[key] = mean(value)

# pp = pprint.PrettyPrinter(depth=4)
# pp.pprint(utilization_altruistic_means)

plt.plot(a1b0_means.values(), label="utilization a=1.0,b=0.0")
plt.legend(loc='best')
plt.show()

plt.plot(a1b25_means.values(), label="utilization a=0.0,b=0.25 (selfish)")
plt.legend(loc='best')
plt.show()

plt.plot(a1b50_means.values(), label="utilization a=0.0,b=0.5")
plt.legend(loc='best')
plt.show()

plt.plot(a1b75_means.values(), label="utilization a=0.0,b=0.75")
plt.legend(loc='best')
plt.show()

plt.plot(a1b100_means.values(), label="utilization a=0.0,b=1.0")
plt.legend(loc='best')
plt.show()

print "Mean utilization for a=0.0,b=0.0 (altruistic) is " + str(mean(a1b0_means.values()))
print "Mean utilization for a=0.0,b=1.0 (selfish) is " + str(mean(a1b25_means.values()))
print "Mean utilization for a=0.0,b=0.9 is " + str(mean(a1b50_means.values()))
print "Mean utilization for a=0.0,b=0.5 is " + str(mean(a1b75_means.values()))
print "Mean utilization for a=0.0,b=0.5 is " + str(mean(a1b100_means.values()))
```
